```python
import pandas as pd
import re
```

```python
f = open('WhatsApp_Chat.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
type(data)
```

```python
pattern = r"(\d{1,2}/\d{1,2}/\d{2,4}), (\d{1,2}:\d{2}\s?[APap][Mm]) - (.*)"
matches = re.findall(pattern, data)
df = pd.DataFrame(matches, columns=["Date", "Time", "Message"])
```

```python
df["DateTime"] = df["Date"] + " " + df["Time"]
df = df[["DateTime", "Message"]]
```

```python
df
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['Message'] :
    entry = re.split('([\w,\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['users'] = users
df['message'] = messages
df.drop(columns=['Message'],inplace=True)

df.head()
```

```python
df["DateTime"] = pd.to_datetime(df["DateTime"], format="%d/%m/%y %I:%M %p")
```

```python
df['year'] = df['DateTime'].dt.year
```

```python
df['month'] = df['DateTime'].dt.month_name()
```

```python
df['day'] = df['DateTime'].dt.day
```

```python
df['hour'] = df['DateTime'].dt.hour
```

```python
df['minute'] = df['DateTime'].dt.minute
```

```python
df['month_num'] = df['DateTime'].dt.month
```

```python
timeline = df.groupby(['year','month_num','month']).count()['message'].reset_index()
```

```python
time = []
for i in range(timeline.shape[0]):
    time.append(timeline['month'][i] + '-' + str(timeline['year'][i]))
```

```python
timeline['time'] = time
```

```python
timeline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(timeline['time'],timeline['message'])
plt.xticks(rotation = 'vertical')
plt.show()
```
